```python
def f(x):
    if x>=7:
        return 0
    else:
        return x
```

```python
t2['Previous'] = t2['Mag'].apply(f)
t2
```

```python
t_array = t2[['Mag', 'Days', 'Previous']].to_numpy()
#t_array2 = t[t['Mag']>=7]['Mag'].to_numpy()
#t_array = t_array[0:1000]
np.shape(t_array)
```

```python
for (i, iq) in enumerate(t_array):
    j = i
    if t_array[i][0]>=7:
        t_array[i][2] = 0
        continue
    while j >= 0:
        j = j-1
        if t_array[j][0] >=7:
            t_array[i][2] = t_array[i][1]-t_array[j][1]
            break
    else:
        t_array[i][2] = t_array[i][1]-t_array[0][1]
np.set_printoptions(suppress = True)
print(t_array)
```
